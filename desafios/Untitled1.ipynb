```python
from astropy.utils.data import get_pkg_data_filename
from astropy.io import fits
from astropy.wcs import WCS
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
img_fname = get_pkg_data_filename('residu_ugc05840.fits')
img_hdu = fits.open(img_fname)[0]
wcs = WCS(img_hdu.header)
#img_hdu.data[(img_hdu.data < 15) & (img_hdu.data > -15)] = np.nan
```

```python
plt.rcParams['figure.figsize'] = 10,10
ALPHA = 1
COLORS = ['blue', 'red']
LEVELS = [-10,10]
plt.subplot(projection=wcs)
#plt.imshow(img_hdu.data, origin='lower',cmap='jet',interpolation='bilinear')
#plt.colorbar(fraction=0.0467, pad=0.02)
plt.contour(img_hdu.data,
            levels=LEVELS,
            #cmap='jet',
            colors=COLORS,
            alpha=ALPHA)
plt.tight_layout()
```

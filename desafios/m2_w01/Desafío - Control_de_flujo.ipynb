```python
import numpy as np
```

```python
a, b, c = 1, 50, 150
ar1 = np.linspace(a, b, num = b-a, dtype=int)
ar2 = np.linspace(b, c, num = c-b, dtype=int)
```

## Desafío 2

```python
for i in ar1:
    if i % 2 == 0:
        print("{} es par".format(i))
    else:
        print("{} es impar".format(i))
```

## Desafío 3

```python
ct2or3, ct2and3, ct3, ct_none = 0, 0, 0, 0
for i in ar2:
    div_2 = i % 2 == 0
    div_3 = i % 3 == 0
    if div_2 | div_3:
        print("{} es divisible por 2 o por 3.".format(i))
        ct2or3 += 1
    if div_2 & div_3:
        print("{} es divisible por 2 y por 3.".format(i))
        ct2and3 += 1
    if div_3 & (not div_2):
        print("{} es divisible por 3 pero no por 2.".format(i))
        ct3 += 1
    if (not div_2) & (not div_3):
        print("{} no es divisible por 2 ni por 3.".format(i))
        ct_none += 1
print("")
print("La condición número 1 se cumple {} veces.".format(ct2or3))
print("La condición número 2 se cumple {} veces.".format(ct2and3))
print("La condición número 3 se cumple {} veces.".format(ct3))
print("La condición número 4 se cumple {} veces.".format(ct_none))
```

## Desafío 6

```python
mean = np.mean(df["passengers"]); print(mean)
```

```python
stdd = np.std(df["passengers"]); print(stdd)
```

```python
df["outlier"] = 0
```

```python
for index, row in df.iterrows():
    is_too_right = (row["passengers"] >= mean + stdd)
    is_too_left = (row["passengers"] <= mean - stdd)
    if is_too_right or is_too_left:
        df.at[index,"outlier"] = 1
```

**Los meses que se escapan de la tendencia son**

```python
df_outliers = df[df["outlier"]==1]["month"].value_counts(); print(df_outliers)
```

**El total de observaciones que se pueden clasificar como casos extremos está dado por**

```python
df_outliers.sum()
```

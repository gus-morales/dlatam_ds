#### Ejercicio 2

Primero veamos qué hace la función actual:

```python
def binarize_histogram(dataframe, variable):
    tmp = dataframe
    tmp['binarize'] = np.where(tmp[variable] > np.mean(tmp[variable]), 1, 0)
    hist_1 = tmp[tmp['binarize'] == 1][variable].dropna()
    hist_0 = tmp[tmp['binarize'] == 0][variable].dropna()
    plt.subplot(1, 2, 1)
    plt.hist(hist_0, alpha=.6, color='lightgrey')
    plt.axvline(np.mean(hist_0))
    plt.title("{0} <= {1}".format(variable, round(np.mean(hist_0), 3)))
    plt.subplot(1, 2, 2)
    plt.hist(hist_1, alpha=.6, color='lightgrey')
    plt.axvline(np.mean(hist_1))
    plt.title("{0} >= {1}".format(variable, round(np.mean(hist_0), 3)))
```

```python
plt.rcParams['figure.figsize'] = 6, 3
```

```python
binarize_histogram(df, vars_list[0])
plt.tight_layout()
```

```python
binarize_histogram(df, vars_list[1])
plt.tight_layout()
```

```python
binarize_histogram(df, vars_list[2])
plt.tight_layout()
```

Una forma de refactorizarla usando `seaborn` puede ser:

```python
def binarize_histogram_sns(dataframe, variable):
    tmp = dataframe
    tmp['bin'] = np.where(tmp[variable] > np.mean(tmp[variable]), 1, 0)
    f, axes = plt.subplots(1, 2, figsize=(7, 3))
    hists = [tmp[tmp['bin'] == i][variable].dropna() for i in np.arange(0, 2)]
    m = round(np.mean(hists[0]), 3)
    strings = ["{0} <= {1}".format(variable, m), "{0} >= {1}".format(variable, m)]
    for ax, h, st in zip(axes, hists, strings):
        sns.distplot(h, kde=False, ax=ax, hist_kws=dict(alpha=.6, color='lightgrey'), bins=10)
        ax.axvline(np.mean(h))
        ax.set_title(st)
        ax.set_xlabel(None)
    plt.tight_layout()
```

```python
for var in vars_list:
    binarize_histogram_sns(df, var)
```

El resultado es idéntico, pero notar que tuve que forzar el binning; mientras `matplotlib` toma el valor de ` rcParams["hist.bins"] = 10`, `seaborn` es relativamente más sofisticado y por defecto usa la regla de *Freedman-Diaconis*, `'fd'`. La función `plt.hist()` de `matplotlib` soporta dicha opción simplemente especificando `bins='fd'`, pero supuse que no podía cambiar la función original del enunciado.

Una segunda forma puede ser:

```python
def binarize_histogram_sns_2(dataframe, variable):
    tmp = dataframe
    tmp['bin'] = np.where(tmp[variable] > np.mean(tmp[variable]), 1, 0)
    grid = sns.FacetGrid(tmp, col='bin', dropna=True, sharex=False, sharey=False)
    grid = grid.map(sns.distplot, variable, kde=False, color='grey', bins=10)
    hists = [tmp[tmp['bin'] == i][variable].dropna() for i in np.arange(0, 2)]
    m = round(np.mean(hists[0]), 3)
    strings = ["{0} <= {1}".format(variable, m), "{0} >= {1}".format(variable, m)]
    axes = grid.axes.flat
    for ax, h, st in zip(axes, hists, strings):
        ax.axvline(np.mean(h))
        ax.set_title(st)
        ax.set_xlabel(None)
```

```python
for var in vars_list:
    binarize_histogram_sns_2(df, var)
```

Y nuevamente obtenemos el mismo resultado.

Ahora veamos qué hace la segunda función:

```python
def grouped_boxplot(dataframe, variable, group_by):
    tmp = dataframe
    stratify_by = tmp[group_by].unique()
    if len(stratify_by) / 2 > 3:
        fig, ax = plt.subplots(2, len(stratify_by),sharey=True)
    else:
        fig, ax = plt.subplots(1, len(stratify_by),sharey=True)
    for i, n in enumerate(stratify_by):
        ax[i].boxplot(tmp[tmp[group_by] == n][variable].dropna())
        ax[i].set_title(n)
```

```python
for var in vars_list:
    grouped_boxplot(df, var, 'region')
plt.tight_layout()
```

Podemos refactorizarla de la siguiente forma:

```python
def grouped_boxplot_sns(dataframe, variable, group_by):
    sns.boxplot(x=df[group_by],y=df[variable].dropna(), width=0.2, color=None, linewidth=1.0)
```

```python
plt.rcParams['figure.figsize'] = 15, 5
plt.style.use('seaborn-muted')
```

```python
for i, var in enumerate(vars_list):
    plt.subplot(1,len(vars_list),i+1)
    grouped_boxplot_sns(df, var, 'region')
plt.tight_layout()
```

Y obtenemos los mismos resultados salvo diferencias de estilo.

Veamos qué hace la última función:

```python
def grouped_scatterplot(dataframe, x, y, group_by):
    tmp = dataframe
    stratify_by = tmp[group_by].unique()
    if len(stratify_by) / 2 > 3:
        fig, ax = plt.subplots(2, len(stratify_by),sharey=True)
    else:
        fig, ax = plt.subplots(1, len(stratify_by),sharey=True)
    for i, n in enumerate(stratify_by):
        tmp_group_plt = tmp[tmp[group_by] == n]
        ax[i].plot(tmp_group_plt[x], tmp_group_plt[y], 'o')
        ax[i].set_title(n)
```

```python
plt.rcParams['figure.figsize'] = 10, 3
```

```python
grouped_scatterplot(df, 'femlab', 'school', 'region')
```

```python
grouped_scatterplot(df, 'femlab', 'literacy', 'region')
```

```python
grouped_scatterplot(df, 'school', 'literacy', 'region')
```

Para refactorizarla, podemos escribir:

```python
def grouped_scatterplot_sns(dataframe, x_var, y_var, group_by):
    sns.lmplot(x=x_var, y=y_var, data=dataframe, hue=group_by, col=group_by, aspect=.5, fit_reg = False)
```

```python
grouped_scatterplot_sns(df, 'femlab', 'school', 'region')
```

```python
grouped_scatterplot_sns(df, 'femlab', 'literacy', 'region')
```

```python
grouped_scatterplot_sns(df, 'school', 'literacy', 'region')
```

Y nuevamente obtenemos los mismos resultados salvo diferencias de formato.

#### Ejercicio 3

```python
plt.rcParams['figure.figsize'] = 7, 5
dfcorr = df.drop(columns=['bin', 'binarize']).corr().dropna()
ax = sns.heatmap(dfcorr[((dfcorr>.6)|(dfcorr<-.6)) & (dfcorr!=1)],
                 annot=True, cmap='viridis',
                 vmin=-1, vmax=1, fmt='.1g')
plt.grid()
```

**(R)** Para un umbral de `R=0.6`:

- `femlab`   → Tasa entre hombres y mujeres en el mercado laboral: **no hay correlación**.
- `school`   → Promedio años de escolaridad se correlaciona:
 * **proporcionalmente** con la *esperanza de vida al nacer* `life` y aún más con la *tasa de alfabetismo* `literacy`;
 * **inversamente proporcional** con la *fertilidad adolescente* `adfert` y aún más con la *mortalidad infantil* `chldmort`.
 
Las correlaciones tienen sentido: no sorprende que una nación con mayor **promedio de años de escolaridad** tenga mayor **esperanza de vida al nacer** (probablemente por mayor acceso a la educación y salud); y debería tener considerablemente mayor **tasa de alfabetismo**, porque naturalmente la escolaridad y el alfabetismo están íntimamente relacionadas. Y en general, tiene sentido que la **fertilidad adolescente** y la **mortalidad infantil** afecten negativamente el acceso a la educación, y por lo tanto el **promedio de años de escolaridad**.

#### Ejercicio 4

```python
tmp = df.loc[:,['region','school','life', 'literacy', 'adfert', 'chldmort']]
regions = df['region'].unique()
```

```python
for reg in regions:
    subset = tmp[tmp['region']==reg].dropna()
    ax = plt.axes()
    ax.set_title(reg)
    sscorr = subset.corr()
    heat = sns.heatmap(sscorr[((sscorr>.6)|(sscorr<-.6)) & (sscorr!=1)],
                 annot=True, cmap='viridis',
                 vmin=-1, vmax=1, fmt='.1g')
    plt.grid()
    plt.show()
```

**Notar que estoy aproximando los números a 1 decimal.**

**(R)**

En general las correlaciones tienen sentido. Señalando algunas:

- Africa:
 * El promedio de años de escolaridad se correlaciona positivamente con la tasa de alfabetismo, y negativamente con la mortalidad infantil.
 * La esperanza de vida al nacer se correlaciona negativamente con la fertilidad adolescente, y aún más con la mortalidad infantil.
 * A su vez, el alfabetismo se correlaciona negativamente con la mortalidad infantil.
 * Finalmente, la fertilidad adolescente se correlaciona positivamente con la mortalidad infantil.
- Americas:
 * Hay una correlación positiva entre el alfabetismo y la escolaridad, y el alfabetismo y la esperanza de vida al nacer.
 * Hay una fuerte correlación negativa entre la mortalidad infantil y la esperanza de vida, y la mortalidad infantil y el alfabetismo.
- Asia:
 * Hay una fuerte correlación negativa entre la esperanza de vida y la mortalidad infantil, y una correlación positiva entre escolaridad y alfabetismo.
- Europe:
 * Hay una correlación negativa entre esperanza de vida y fertilidad adolescente, y esperanza de vida y mortalidad infantil. También hay una correlación positiva entre fertilidad adolescente y mortalidad infantil.
- Oceania:
 * Fuerte correlación negativa entre esperanza de vida y fertilidad adolescente, y esperanza de vida y mortalidad infantil.
 * Fuerte correlación positiva entre escolaridad y alfabetismo, y fertilidad adolescente y mortalidad infantil.

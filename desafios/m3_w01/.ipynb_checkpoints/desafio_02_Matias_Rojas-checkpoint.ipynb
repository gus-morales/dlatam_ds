### Ejercicio 3: Modelo

```python
from sklearn.model_selection import train_test_split
from pygam import LinearGAM #dado que el vector objectivo es continuo
from pygam.utils import generate_X_grid 
from sklearn.metrics import r2_score, mean_squared_error, median_absolute_error
```

```python
x = df.drop(['target'], axis=1)  # features
y = df['target']  # target
```

**Conjuntos de entrenamiento y validación.**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.33, random_state=63)
```

**GAM sin gridsearch con un lambda de 1000**

```python
gam1000 = LinearGAM(lam = 1e3) #lambda = 1000
gam1000.fit(x_train, y_train)
```

**lambda fijo**

```python
gam1000.lam
```

**Summary Gam1000**

```python
gam1000.summary()
```

**GAM con gridsearch**

```python
#search_params = {'lam': np.logspace(-3,3,11)}, no se incorpora el parametro, dado que según la documentación
#cuando no está especificado se pone el mismo valor -3,3,11 por defecto.

gam = LinearGAM().gridsearch(x_train, y_train)
```

**lambda óptimo**

```python
gam.lam
```

**summary**

```python
gam.summary()
```

### Ejercicio 4: Dependencia parcial

```python
attribute = x_train.columns #nombre de las columnas
cols=8; #columnas de la grilla
rows = int(len(attribute)/cols) #determina la cantidad de filas
```

**Gam1000 sin gridsearch**

```python
x_grid = generate_X_grid(gam1000)
for i, n in enumerate(range(len(attribute))):
    plt.subplot(rows, cols, i + 1)
    partial_dep, confidence_intervals = gam1000.partial_dependence(x_grid, feature=i+1, width=.95) #evalua la funcion de identidad para cada atributo especifico
    plt.plot(x_grid[:, n], partial_dep, color='tomato')
    plt.fill_between(x_grid[:, n],
        confidence_intervals[0][:, 0],
        confidence_intervals[0][:,1],
        alpha=.25)
    plt.title(attribute[n])
    plt.scatter(x_train[attribute[n]], [plt.ylim()[0]] * len(x_train[attribute[n]]), marker='|', alpha=.5)
plt.tight_layout()
```

**Gam con gridsearch**

```python
x_grid = generate_X_grid(gam)
```

```python
for i, n in enumerate(range(len(attribute))): #recorre cada columna del vector x_train
    plt.subplot(rows, cols, i + 1)
    partial_dep, confidence_intervals = gam.partial_dependence(x_grid, feature=i+1, width=.95) #evalua la funcion de identidad para cada atributo especifico
    plt.plot(x_grid[:, n], partial_dep, color='tomato')
    plt.fill_between(x_grid[:, n],
            confidence_intervals[0][:, 0],
            confidence_intervals[0][:,1],alpha=.25) #Genera intervalos de confianza superior e inferior
    plt.title(attribute[n])
    plt.scatter(x_train[attribute[n]], [plt.ylim()[0]] * len(x_train[attribute[n]]), marker='|', alpha=.5)
plt.tight_layout()
```

**eje y : representa valor predicho a lo largo que atributo aumenta de valor**
* Cement: a medida que la cantidad de cemento, mayor resistencia a la compresión del hormigón
* cuando Blast Furnace Slag, supera a 200 la resistencia a la compresión es mayor, en valores elevadoss de blast furnace, la resistencia disminuye
* flash ash: a mayor cantidad de cenizas menor resistencia del hormigón
* water: a menor cantidad de agua mayor es la resistencia del hormigón
* superplasticizer: a mayor cantidad de plastificante la resitencia del hormigón tiende a disminuir
* coarse: a mayor espesor del hormigón mayor es su resistencia
* agregado fino:  a mayor agregado fino, mejor es la resitencia del hormigón
* Age: a mayor cantidad de días el hormigón tiende a disminuir su resitencia.

**Métricas comparativas**

```python
R2= [(r2_score(gam1000.predict(x_test), y_test)),
     (r2_score(gam.predict(x_test), y_test))]
MSE= [(mean_squared_error(gam1000.predict(x_test), y_test)),
       (mean_squared_error(gam.predict(x_test), y_test))]
MAE= [(median_absolute_error(gam1000.predict(x_test), y_test)),
      (median_absolute_error(gam.predict(x_test), y_test))]
```

```python
metricas= pd.DataFrame(data={
                   'R2': R2,
                   'MSE': MSE,
                   'MAE': MAE},
             index=['GAM1000','GAM']).transpose()
```

También, usaré algunas funciones de la vez pasada:

```python
def inverse_logit(x):
    return 1 / (1 + np.exp(-x))
```

```python
def generate_fl(variables, dep_variable):
    string = dep_variable + ' ~ '
    for var in variables:
        if var != dep_variable:
            string += var
            string += ' + '
    string = string[:-3]
    return string
```

```python
def compute_log_probability(dataframe, model, variables, x):
    logodds = 0
    for var in variables:
        if var[:3] != 'is_':
            logodds += model.params[var]*dataframe[var].mean()
        else:
            logodds += model.params[var]*x + model.params[var]*0
    logodds += model.params['Intercept']
    return inverse_logit(logodds)
```

```python
df = df.drop(columns=['student', 'default'])
```

```python
generate_fl(df.columns.tolist(), 'is_Defaulter')
```

Y puedo generar mi modelo saturado de la siguiente forma:

```python
mod_logit_sat = smf.logit(generate_fl(df.columns.tolist(), 'is_Defaulter'), df).fit()
```

```python
mod_logit_sat.summary2()
```

Una significancia de $95\%$ corresponde a un p-value de $5\%$, es decir:

```python
CRIT_VALUE = 5/100
```

```python
table = mod_logit_sat.summary2().tables[1]['P>|z|']
```

```python
high_sig = [ table.index[i] for i, value in enumerate(table) if value <= CRIT_VALUE and i != 0 ]
```

```python
string2 = 'is_Defaulter ~ '
for var in high_sig:
    if var != 'is_Defaulter':
        string2 += var
        string2 += ' + '
string2 = string2[:-3]
print(string2)
```

```python
mod_logit_high_sig = smf.logit(string2, df).fit()
```

```python
mod_logit_high_sig.summary2()
```

Podemos ver que el efecto de ser estudiante es significativamente mayor que el saldo en la cuenta.

Para estimar la probabilidad de ser moroso,

```python
p1 = compute_log_probability(df, mod_logit_high_sig, high_sig, 1)
```

```python
p0 = compute_log_probability(df, mod_logit_high_sig, high_sig, 0)
```

```python
print(f"La probabilidad de entrar en morosidad de individuo:\n\
 - con un total de saldo similar al promedio de la muestra;\n\
 - que ES estudiante;\n\
es de un {p1*100:.2f}%")
```

```python
print(f"La probabilidad de entrar en morosidad de individuo:\n\
 - con un total de saldo similar al promedio de la muestra;\n\
 - que NO es estudiante;\n\
es de un {p0*100:.2f}%")
```

Revisemos la _confusion matrix_:

```python
cm = confusion_matrix(y_test, y_test_predict)
tn, fp, fn, tp = cm.ravel()
df1 = pd.DataFrame.from_dict({'predicted positive': f'TP = {tp}', 'predicted negative': f'FN = {fn} (type II)'},
                             orient='index', columns=['actual positive'])
df2 = pd.DataFrame.from_dict({'predicted positive': f'FP = {fp} (type I)', 'predicted negative': f'TN = {tn}'},
                             orient='index', columns=['actual negative'])
pd.concat([df1,df2], axis=1)
```

```python
def get_scores(cm_object):
    tn, fp, fn, tp = cm_object.ravel()
    precision = tp / (tp + fp)
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    neg_pred_value = tn / (tn + fn)
    f1 = 2*(precision * sensitivity)/(precision + sensitivity)
    """
    precision (positive predictive value): probability that instances predicted positive are actually positive → a measure of QUALITY
    accuracy: fraction of correct predictions among the total number of input samples
    sensitivity (recall): how many of the actual positives our model capture through labeling it as positive → a measure of QUANTITY (completeness)
    neg_pred_value: probability that instances predicted negative are actually negative
    f1: (precision plus recall) a combined measure of how many instances are classified correctly,
        as well as how robust it is by not missing a significant number of instances → a measure of TEST'S ACCURACY
    """
    return {'precision': precision,
            'accuracy': accuracy,
            'sensitivity': sensitivity,
            'specificity': specificity,
            'neg_pred_value': neg_pred_value,
            'f1': f1}
```

```python
get_scores(cm)
```

Cuando consideramos los resultados de un test en particular para dos poblaciones, en este caso una de morosos y otra de no morosos, rara vez el test va a ser capaz de separar correctamente entre estos dos grupos, debido a errores Tipo I (falsos positivos) y Tipo II (falsos negativos).

<img src="roc_intro2.png">

Para cada caso posible de corte que se seleccione para discriminar entre las dos muestras, algunos casos positivos van a ser identificados correctamente (TP), pero otros casos negativamente (FN). Similarmente, algunos casos realmente negativos serán predichos como negativos (TN), y otros como positivos (FP). Cambiando el umbral aumenta efectivamente la tasa de TP o TN, pero **a costa** de aumentar también la tasa de FP y FN. Se puede entonces encontrar un umbral crítico que minimize la cantidad de FP y FN; es decir, que sea capaz de separar lo más posible ambas poblaciones. Este problema es lo que busca representar la ROC curve.

Se definen entonces:

`sensitivity` o `recall`, como la probabilidad de que un test resulte positivo cuando realmente lo es: `TP / (TP + FN)`; y

`specificity`, como la probabilidad de que un test resulte negativo cuando realmente lo es `TN / (TN + FP)`. En consecuencia, `1 - specificity`, es simplemente la _false positive rate_, o tasa de falsos positivos.

En una curva ROC se grafica la `sensitivity` (_true positive rate_) como función de `1 - specificity` (_false positive rate_), para distintos puntos de corte de cierto parámetro. Así, cada punto en una curva ROC representa un par `sensitivity`/`specificity` correspondiente a un umbral de decisión en particular. Entonces, un test que es capaz de discriminar perfectamente entre ambas poblaciones (sin nada de _overlap_) debe tener una curva ROC que pasa por el extremo superior izquierdo del gráfico, es decir, 100% `sensitivity` y 100% `specificity` (o 0% de `1 - specificity`). En la práctica, esto se traduce a que el área bajo la curva ROC (AUC) es una buena medida de cuán bien se pueden discriminar ambas poblaciones, reduciendo los errores Tipo I y II. Un AUC de 100% quiere decir que para cierto umbral, el modelo es capaz de separar de forma perfecta ambas poblaciones sin errores de ningún tipo.

```python
auc = roc_auc_score(y_test, y_test_predict_probs)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_test_predict_probs)
```

```python
plt.rcParams['figure.figsize'] = 6, 4
plt.plot(fpr, tpr, label=f'ROC curve: AUC={auc:0.2f}')
plt.xlabel('1 - Specificity (false positive rate)')
plt.ylabel('Sensitivity (true positive rate)')
plt.grid(True)
plt.legend()
plt.tight_layout()
```

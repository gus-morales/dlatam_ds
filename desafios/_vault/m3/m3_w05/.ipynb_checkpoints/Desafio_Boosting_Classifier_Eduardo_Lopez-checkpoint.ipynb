```python
def plot_describe_variables(dataframe):
    plt.rcParams['figure.figsize'] = 25, 20
    for index, (colname, serie) in enumerate(dataframe.iteritems()):
        plt.subplot(5, 4, index + 1)
        if serie.dtype == 'object':
            sns.countplot(serie)
            plt.axhline(serie.value_counts().mean())
            plt.title(colname)
        else:
            sns.distplot(serie)
            plt.axvline(np.mean(serie))
            plt.title(colname)
    plt.tight_layout()
```

## Comportamiento distributivo para cada atributo 

```python
plot_describe_variables(df)
```

#### Descripción del comportamiento de atributos:
En general las variables presentan un comportamiento normal. Sin embargo existen 3 atributos continuos que presentan alto sesgo por lo que se procedera a convertir a logaritmo, siendo estos: 
* number_vmail_messages
* number_customer_service_calls
* total_intl_calls  

Además por fines practicos de modelamiento se procedera a convertir las variables de tipo discreto o clases a binario( 0, 1).

Para variables Discretas:

```python
col_del = []
for i, col in enumerate(df):
    if df[col].dtype == "object":
        df = pd.concat([df, pd.get_dummies(df[col],
                                           drop_first = True, 
                                           prefix = col)], 
                       axis = 1)
        col_del.append(col)
        
df = df.drop(columns = col_del)
```

```python
df
```

Para las 3 variables continuas se procesan a valor logaritmico:

```python
log_vars = ["number_vmail_messages",
            "number_customer_service_calls",
            "total_intl_calls"]

for i in log_vars:
    df[i] = df[i].apply(lambda x: np.log(x + 0.01))
```

# Ejercicio 2: Comparación de AdaBoost y Gradient Boosting
* Entrene los clasificadores AdaBoost y Gradient Boosting para mejorar su capacidad predictiva en la medida de lo posible. Para ello, implemente una búsqueda de grilla con las siguientes especificaciones:
* Si el tiempo de computación es alto, puede implementar la búsqueda con 1 validación cruzada. Reporte las métricas para los mejores modelos.

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
```

```python
X = df.drop(["churn_yes"], axis = 1)
y = df["churn_yes"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .33, random_state = 63)
```

## AdaBoostClassifier mediante GridSearch

```python
ada_params = {'learning_rate': [0.01, 0.1, 0.5],
              'n_estimators': [50, 100, 500, 1000, 2000]}

grid_ada_model = GridSearchCV(AdaBoostClassifier(random_state = 63), ada_params,
                             cv = 2).fit(X_train, y_train)
```

### Mejor seleccion de hiperparametros para Ada Boost Classifier

```python
print(grid_ada_model.best_params_)
```

```python
pred_ada = grid_ada_model.predict(X_test)
```

```python
print(classification_report(y_test, pred_ada))
```

### El modelo mediante grilla utilizando AdaBoost presenta un desempeño promedio en su predicción mediante f1-score del 88%.

## GradientBoostingClassifier mediante GridSearch

```python
gradient_params = {'learning_rate': [0.01, 0.1,0.5],
                   'n_estimators': [50, 100, 500, 1000, 2000],
                   'subsample': [0.1,0.5,0.9]}

grid_gradient_model = GridSearchCV(GradientBoostingClassifier(), gradient_params,
                             cv = 2).fit(X_train, y_train)
```

### Mejor seleccion de hiperparametros para Gradient Boosting Classifier

```python
print(grid_gradient_model.best_params_)
```

```python
pred_gradient = grid_gradient_model.predict(X_test)
```

```python
print(classification_report(y_test, pred_gradient))
```

### El modelo mediante grilla utilizando AdaBoost presenta un desempeño promedio en su predicción mediante f1-score del 95%.
### Además presenta una precision importante para la prediccion tanto de 0 como 1.

# Ejercicio 3: Principales factores asociados
* Con el mejor modelo, reporte la importancia relativa de los atributos y comente cuáles son los que aumentan la probabilidad de fuga en los clientes.
* tip: Pueden implementar la función plot_importance que se encuentra en la lectura de Bagging y Random Forest.

```python
def plot_importance(fit_model, feat_names):
    """TODO: Docstring for plot_importance.

    :fit_model: TODO
    :: TODO
    :returns: TODO

    """
    tmp_importance = fit_model.best_estimator_.feature_importances_ # para acceder a features importantes
    sort_importance = np.argsort(tmp_importance)[::-1]
    names = [feat_names[i] for i in sort_importance]
    plt.title("Feature importance")
    plt.barh(names[:20], tmp_importance[sort_importance][:20])
    plt.yticks(names[:20], names, rotation = 0)
```

```python
plt.rcParams['figure.figsize'] = 10, 10
plot_importance(grid_gradient_model, X.columns)
```

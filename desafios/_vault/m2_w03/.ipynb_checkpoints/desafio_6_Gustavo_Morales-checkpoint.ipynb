### Ejercicio 3

```python
def compute_zscore(dataframe, variable):
    mean_var = dataframe[variable].mean()
    stdd_var = dataframe[variable].std(ddof=0)  # to handle nan values
    z_score = (dataframe[variable] - mean_var)/stdd_var
    return z_score
```

Sin considerar las columnas sin NaN's, utilizaré las siguientes columnas:

```python
least_na_fraction = ['gle_cgdpc', 'wdi_exph', 'undp_hdi']
```

```python
for var in least_na_fraction:
    df['z_'+var] = pd.Series(compute_zscore(df, var))
```

```python
df = df.dropna(subset=least_na_fraction)
```

Compruebo que `df` ya no tenga valores NaN's:

```python
pd.DataFrame([af.get_na(df, var) for var in least_na_fraction]).sort_values(by='na_fraction')
```

Para analizar los primeros y últimos 10 registros:

```python
from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
s_head = [df['z_'+var].sort_values(ascending=True).head(10).reset_index(drop=True) for var in least_na_fraction]
s_tail = [df['z_'+var].sort_values(ascending=True).tail(10).reset_index(drop=True) for var in least_na_fraction]
```

```python
display_side_by_side(pd.DataFrame(s_head).transpose(), pd.DataFrame(s_tail).transpose())
```

**(R) La característica en común es que como el `z_score` normaliza los promedios para las tres variables en `0`, al ordenarlos de menor a mayor los 10 primeros registros son negativos, y los 10 últimos positivos, relativamente equidistantes de cero.**

### Ejercicio 5

```python
fname2 = "qog_std_cs_jan18.csv"
df_tot = pd.read_csv(fname2)
```

```python
def _aux(dfr, var, log, c, a, lw, leg):
    dfr_obj = np.log10(dfr[var].dropna()) if log else dfr[var].dropna()
    x = dfr_obj.values.tolist()
    gkde = gaussian_kde(x)
    ind = np.linspace(min(x),max(x),101)
    kdepdf = gkde.evaluate(ind)
    plt.plot(ind, kdepdf, color=c, lw=lw, label=leg)
```

```python
def plot_2kde(df1, df2, variable, log=False):
    ALPHA, LW = 0.3, 3
    C1, C2 = 'tomato', 'dodgerblue'
    leg1, leg2 = 'total', 'sample'
    _aux(df1, variable, log, C1, ALPHA, LW, leg1)
    _aux(df2, variable, log, C2, ALPHA, LW, leg2)
```

```python
vars_to_plot = ['imf_pop', 'gle_cgdpc', 'undp_hdi']
plt.figure(figsize=(14,3))
for i, var in enumerate(vars_to_plot):
    plt.subplot(1,len(vars_to_plot),i+1)
    plot_2kde(df_tot, df, var, log=False) if i+1==3 else plot_2kde(df_tot, df, var, log=True)
    plt.title("Gaussian KDE's"); plt.legend() if i+1==2 else plt.title('')
    plt.xlabel(var)
plt.tight_layout()
```

**(R) En los dos primeros casos, `imfpop` y `gle_cgdpc`, es útil usar una escala logarítmica porque los datos presentan una asimetría (*skewness*) considerable hacia valores mayores. En general, usar logaritmos en base 10 `np.log10()` es conveniente cuando los datos cubren varios órdenes de magnitud, pero también se pueden usar otras bases.**

Una forma alternativa (y más cómoda) es usar `seaborn.distplot()`:

```python
import seaborn as sns
```

```python
plt.figure(figsize=(14,3))
plt.subplot(1,3,1)
sns.distplot(np.log(df_tot['imf_pop']).dropna())
sns.distplot(np.log(df['imf_pop']).dropna())
plt.xlabel('imf_pop')
plt.subplot(1,3,2)
sns.distplot(np.log(df_tot['gle_cgdpc']).dropna())
sns.distplot(np.log(df['gle_cgdpc']).dropna())
plt.xlabel('gle_cgdpc')
plt.subplot(1,3,3)
sns.distplot(df_tot['undp_hdi'].dropna())
sns.distplot(df['undp_hdi'].dropna())
plt.xlabel('undp_hdi')
plt.tight_layout()
```
